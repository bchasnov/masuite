### Plotting Function

```python
# Given parameters plot graph
def data_plot(fig, ax, x, y, xlabeler, ylabeler, conv_label=False, figsize=[9, 4]):
    
    # Plot
    fig.set_size_inches(figsize)
    ax.plot(x, y, 'go')
    
    # Plot Trendline
    z = np.polyfit(x, y, 1)
    p = np.poly1d(z)
    plt.plot(x, p(x),"r--")

    ax.set_xlabel(xlabeler)
    ax.set_ylabel(ylabeler)
    ax.set_title(xlabeler + ' vs ' + ylabeler + ' in IPD')

    return 0
```

### Graphs

```python
# Plot data
figure
plt.ylabel("Cooperation Rating")
plt.xlabel("Median Score")
plt.xlim([2.1, 2.4])
plt.ylim([0, 1])
plt.title("Cooperating Rating vs Median Score in IPD")
plt.plot(data.Median_score, data.Cooperation_rating, 'go')

# Plot trendline
z = np.polyfit(data.Median_score, data.Cooperation_rating, 1)
p = np.poly1d(z)
plt.plot(data.Median_score,p(data.Median_score),"r--")
```
