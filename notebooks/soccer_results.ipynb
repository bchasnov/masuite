```python
import pandas as pd
import matplotlib.pyplot as plt
import masuite.experiments.smallsoccer.analysis as analysis
```

```python
def plot_rets(title, df):
    plt.title(title)
    plt.plot(df["agent0_avg_rets"], label="player 1")
    plt.plot(df["agent1_avg_rets"], label="player 2")
    plt.legend()
```

```python
simplepg_soccer_path = "/home/zanedma/code/research/masuite/tmp/masuite/SimplePG-soccer-0.csv"
simplepg_df = pd.read_csv(simplepg_soccer_path)
plot_rets("SimplePG Littman Soccer", simplepg_df)
```

```python
stackpg_soccer_path = "/home/zanedma/code/research/masuite/tmp/masuite/StackPG-soccer-0.csv"
stackpg_df = pd.read_csv(stackpg_soccer_path)
plot_rets("StackPG Littman Soccer", stackpg_df)
```

```python
simplepg_ssoccer_path = "/home/zanedma/code/research/masuite/tmp/masuite/SimplePG-smallsoccer-0.csv"
simplepg_sdf = pd.read_csv(simplepg_ssoccer_path)
plot_rets("SimplePG Small Soccer", simplepg_sdf)
```

```python
stackpg_ssocer_path = "/home/zanedma/code/research/masuite/tmp/masuite/StackPG-smallsoccer-0.csv"
stackpg_sdf = pd.read_csv(stackpg_ssocer_path)
plot_rets("StackPG Small Soccer", stackpg_sdf)
```
