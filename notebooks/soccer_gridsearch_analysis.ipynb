```python
import masuite.experiments.smallsoccer.analysis as analysis
import pandas as pd
```

```python
path = "/home/zanedma/code/research/masuite/examples/tmp/masuite"
results = analysis.score_gridsearch(path)
results.head()
```

# Seeded

```python
results.sort_values(by="score")
```

```python
params = results.filename[0].split("_")
print(params)
```

```python
paresed_results = pd.DataFrame(columns=["score", "epochs", "batch size", "lr"])
for res in results.itertuples():
    params = res.filename.split("_")
    epochs = int(params[1].replace("epochs", ""))
    batch_size = int(params[3].replace("size", ""))
    lr = float(params[-2].replace("lr", ""))
    row = [res.score, epochs, batch_size, lr]
    paresed_results.loc[len(paresed_results.index)] = row
paresed_results.sort_values(by="score")
```
